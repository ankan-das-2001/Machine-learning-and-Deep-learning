```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv("titanic/train.csv")
test_data = pd.read_csv("titanic/test.csv")
```

```python
columns_to_drop = ["PassengerId","Name","Ticket","Cabin","Embarked"]
train_data = train_data.drop(columns_to_drop,axis=1)
test_data = test_data.drop(columns_to_drop,axis=1)
```

```python
train_data.head(10)
```

```python
train_data["Age"] = train_data["Age"].fillna(train_data["Age"].mean())
test_data["Age"] = test_data["Age"].fillna(test_data["Age"].mean())
```

```python
test_data.head(10)
```

```python
# label encoding the sex columns males are 1 and female are 0
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
train_data["Sex"] = le.fit_transform(train_data["Sex"])
test_data["Sex"] = le.fit_transform(test_data["Sex"])
```

```python
test_data.head(10)
```

```python
# All the data preprocessing is done now the step is to create out X and y values and fit the prediction
X_labels = ["Pclass","Sex","Age","Fare","SibSp","Parch"]
y_label = ["Survived"]

X_train = train_data[X_labels]
y_train = train_data[y_label]

X_test = test_data[X_labels]
```

```python
# implementing decision tree
# from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
d_tree = RandomForestClassifier(n_estimators=21, criterion="entropy",max_depth=5)

# d_tree = DecisionTreeClassifier(criterion="entropy", max_depth=5)
d_tree.fit(X_train, y_train)
```

```python
d_tree.score(X_train, y_train)
```

```python
# survivor_predictions = d_tree.predict(X_test)
X_test["Fare"] = X_test["Fare"].fillna(X_test["Fare"].mean())
survivor_predictions = d_tree.predict(X_test)
```

```python
final_result = pd.read_csv("titanic/gender_submission.csv")
final_result.head(10)
```

```python
final_result["Survived"]=survivor_predictions
final_result.head(10)
```

```python
final_result.to_csv("Final_result.csv",index=False)
```
