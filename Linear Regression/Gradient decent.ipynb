```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
fx = (X-5)**2
print(X,fx)
```

```python
plt.style.use("seaborn")
plt.plot(X,fx)
plt.xlabel("X")
plt.ylabel("F[x]")
plt.show()
```

```python
# The gradient decent algorithm
#grad is calculated differantiating the fx function
x=0
learning_rate=  0.1
error = []
plt.plot(X,fx)
for i in range(50):
    grad = 2*(x-5)
    x = x - learning_rate*grad
    e = (x-5)**2
    error.append(e)
    plt.scatter(x,e)
plt.show()
#     print(x)
```
