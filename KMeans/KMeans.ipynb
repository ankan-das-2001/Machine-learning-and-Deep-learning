# Implementation of KMeans


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

# Dataset Preparation

```python
X,y = make_blobs(n_samples=500,n_features=2, centers=5,random_state=3)
```

```python
plt.figure(0)
plt.scatter(X[:,0], X[:,1],c=y)
```

```python
k=5
color=["green","red","blue","yellow","orange"]

clusters = {}

for kx in range(k):
    center = 10*(2*np.random.random((X.shape[1],))-1)
    points = []
    cluster = {
        'center':center,
        'points': points,
        'color': color[kx]
    }
    
    clusters[kx] = cluster
```

```python
clusters
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
def assignPointsToCluster(cluster):
    for ix in range(X.shape[0]):
        dist = []
        curr_x = X[ix]

        for kx in range(k):
            d = distance(curr_x, clusters[kx]['center'])
            dist.append(d)

        current_cluster = np.argmin(dist)

        clusters[current_cluster]['points'].append(curr_x)
```

```python
def updateCluster(cluster):   #M steps
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        
        if pts.shape[0]>0:
            new_u = pts.mean(axis=0)
            cluster[kx]['center'] = new_u
            cluster[kx]['points'] = []

def plotCluster(cluster):
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        #plot the point
        try:
            plt.scatter(pts[:,0],pts[:,1],color=clusters[kx]['color'])
        except:
            pass
        #plot the cluster center
        uk=cluster[kx]['center']
        plt.scatter(uk[0],uk[1],color="black")
```

```python
assignPointsToCluster(clusters)
plotCluster(clusters)
updateCluster(clusters)
```
