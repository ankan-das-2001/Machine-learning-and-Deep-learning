```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
pic1 = cv2.imread("top_left.jpg")   
pic2 = cv2.imread("top_right.jpg")
pic3 = cv2.imread('bottom_left.jpg')
pic4 = cv2.imread('bottom_right.jpg')
pic5 = cv2.imread('center.jpeg')
```

```python
pic1 = cv2.resize(pic1,(200,200))
pic2 = cv2.resize(pic2,(200,200))
pic3 = cv2.resize(pic3,(200,200))
pic4 = cv2.resize(pic4,(200,200))
pic5 = cv2.resize(pic5,(100,100))
pic1 = cv2.cvtColor(pic1, cv2.COLOR_BGR2RGB)
pic2 = cv2.cvtColor(pic2, cv2.COLOR_BGR2RGB)
pic3 = cv2.cvtColor(pic3, cv2.COLOR_BGR2RGB)
pic4 = cv2.cvtColor(pic4, cv2.COLOR_BGR2RGB)
pic5 = cv2.cvtColor(pic5, cv2.COLOR_BGR2RGB)
```

```python
# Implementing a matrix
```

```python
matrix = np.zeros((430,430,3),dtype = np.uint8)
```

```python
matrix.shape
```

```python
plt.imshow(matrix)
```

```python
matrix[10:210,10:210,:] = pic1
matrix[10:210, 220:420,:] = pic2
matrix[220:420, 10:210, :] = pic3
matrix[220:420, 220:420, :] = pic4
```

```python
plt.figure(figsize=(12,12))
plt.axis("off")
plt.imshow(matrix)
```

```python
cen_matrix = np.zeros((120,120,3), dtype = np.uint8)
```

```python
cen_matrix[10:110,10:110,:] = pic5
plt.imshow(cen_matrix)
```

```python
matrix[155:275,155:275,:] = cen_matrix
```

```python
plt.figure(figsize=(5,5))
plt.axis("off")
plt.imshow(matrix)
```

```python
# Saving the image
cv2.imwrite("Collage.jpg", cv2.cvtColor(matrix,cv2.COLOR_RGB2BGR))
```
