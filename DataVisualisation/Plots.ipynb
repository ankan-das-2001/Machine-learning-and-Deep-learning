```python
import matplotlib.pyplot as plt
import numpy as np

themes = plt.style.available
print(themes)
```

# Line Plot

```python
plt.style.use("seaborn")

x = np.arange(10)
y1 = x**2
y2 = 2*x+3
print(x,y1,y2)
```

```python
plt.plot(x,y1, color="red", label="Apple", marker="o")
plt.plot(x,y2, color="orange", label="banana",linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of fruits over time")
plt.legend()
plt.show()

# Plots can also be done using 1-d array
```

# Scatter plot

```python
#Adjust the size of any plot
plt.figure(figsize=(10,10))

plt.scatter(x,y1, color="red", label="Apple", marker="^")
plt.scatter(x,y2, color="orange", label="banana",linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot - Prices of fruits over time")
plt.legend()
plt.show()
```

# Bar-graphs

```python

x = np.array([0,1,2])*2
plt.bar(x,[10,20,3],color="orange",width=0.5,label = "current year", tick_label=["gold","silver","platinum"])
plt.bar(x+0.5,[14,10,12],color="red", width=0.5, label="next year")
plt.title("metals price comparion")
plt.xlabel("metals")
plt.ylabel("price")

plt.ylim(0,40)
plt.legend()
plt.show()
```

# Pie Chart


```python
plt.figure(figsize=(10,10))
subjects = ["Maths", "chemistry","English","Physics","Computer"]
weightage = [20,10,15,5,22]
plt.pie(weightage, labels=subjects, explode=(0,0,0,0,0.2), shadow=True, autopct = "%1.1f%%")
plt.show()
```

# Histogram

```python
xsn = np.random.randn(100)
sigma = 13
u=70
X = np.round(xsn*sigma+u)
X2 = np.round(xsn*5+40)
```

```python
print(X)
```

```python
plt.figure(figsize=(10,10))
plt.style.use("seaborn")
plt.hist(X)
plt.hist(X2)
plt.ylabel("Frequency of student")
plt.xlabel("Marks Range")
```
